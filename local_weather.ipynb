```python
def create_predictions(predictors,core_weather, reg):
  pass
  train = core_weather.loc[:"2018-12-31"]
  test = core_weather.loc["2019-11-14":]
  reg.fit(train[predictors],train["Temperature Prediction"])
  predictions = reg.predict(test[predictors])
  combined = pd.concat([test["Temperature Prediction"],pd.Series(predictions,index=test.index)],axis=1)
  combined.columns=["actual","prediction"]
  return combined, mean_absolute_error(train["Temperature Prediction"],reg.predict(train[predictors]))
```

```python
core_weather["month_max"]=core_weather["Temperature"].rolling(30).mean()
core_weather
```

```python
core_weather["month_day_max"]=core_weather["month_max"]/core_weather["Temperature"]
```

```python
predictors=["Temperature","Humidity","month_max","month_day_max"]
```

```python
core_weather = core_weather.iloc[30:,:].copy()
```

```python
error, combined = create_predictions(predictors,core_weather, reg)
```
